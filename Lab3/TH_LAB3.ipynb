```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

def apply_translation(image_data, shift_x, shift_y):

    print(f"Áp dụng tịnh tiến: X={shift_x}, Y={shift_y}")
    transform_matrix = np.array([[1, 0, shift_x],
                                 [0, 1, shift_y],
                                 [0, 0, 1]])
    
    def map_coords(coords):
        row, col = coords[:2]
        if image_data.ndim == 3:
            return (row - shift_y, col - shift_x, coords[2])
        return (row - shift_y, col - shift_x)

    translated_image = nd.geometric_transform(image_data, map_coords, 
                                             output_shape=image_data.shape, 
                                             order=1, mode='constant', cval=0)
    return translated_image

def apply_rotation(image_data, angle_degrees, reshape=True):
    """
    Thực hiện xoay ảnh một góc.
    angle_degrees: góc xoay theo độ.
    reshape: Nếu True, thay đổi kích thước ảnh để chứa toàn bộ ảnh đã xoay.
             Nếu False, giữ nguyên kích thước, các phần ảnh ngoài biên sẽ bị cắt.
    """
    print(f"Áp dụng xoay: {angle_degrees} độ, reshape={reshape}")
    rotated_image = nd.rotate(image_data, angle_degrees, reshape=reshape, order=1, cval=0)
    return rotated_image

def apply_zoom(image_data, factor):
    """
    Phóng to hoặc thu nhỏ ảnh.
    factor: Hệ số zoom (factor > 1 cho phóng to, factor < 1 cho thu nhỏ).
    """
    print(f"Áp dụng zoom: {factor} lần")
    if image_data.ndim == 3:
        zoomed_image = nd.zoom(image_data, (factor, factor, 1), order=1)
    else:
        zoomed_image = nd.zoom(image_data, factor, order=1)
    return zoomed_image

def apply_coordinate_map(image_data, d=5):
    """
    Áp dụng biến đổi Coordinate Map (hiệu ứng ngẫu nhiên nhẹ).
    d: biên độ dịch chuyển ngẫu nhiên.
    """
    print(f"Áp dụng Coordinate Map với d={d}")
    V, H = image_data.shape[:2]
    
    M = np.indices((V, H))
    
    q = (2 * d * np.random.rand(*M.shape) - d)
    
    mp = (M + q)
    if image_data.ndim == 3:

        mapped_image = np.zeros_like(image_data)
        for i in range(image_data.shape[2]):
            mapped_image[:,:,i] = nd.map_coordinates(image_data[:,:,i], mp, order=1, mode='reflect')
    else:
        mapped_image = nd.map_coordinates(image_data, mp, order=1, mode='reflect')
        
    return mapped_image.astype(image_data.dtype)

def main_menu():
    images = {
        '1': 'colorful-ripe-tropical-fruits.jpg',
        '2': 'pagoda.jpg',
        '3': 'quang_ninh.jpg'
    }

    while True:
        print("\n--- Chọn hình ảnh ---")
        for key, value in images.items():
            print(f"{key}. {value}")
        print("0. Thoát")

        image_choice = input("Chọn số của hình ảnh bạn muốn (1, 2, 3) hoặc 0 để thoát: ").strip()

        if image_choice == '0':
            print("Thoát chương trình. Tạm biệt!")
            break
        
        if image_choice not in images:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
            continue

        selected_image_path = images[image_choice]
        try:
            current_image_data = iio.imread(selected_image_path)
            print(f"Đã tải ảnh: {selected_image_path} với kích thước {current_image_data.shape}")
        except FileNotFoundError:
            print(f"Lỗi: Không tìm thấy file ảnh '{selected_image_path}'. Vui lòng đảm bảo nó nằm trong cùng thư mục với script.")
            continue
        except Exception as e:
            print(f"Lỗi khi đọc ảnh: {e}")
            continue

        while True:
            print(f"\n--- Thao tác trên {selected_image_path} ---")
            print("T. Tịnh tiến")
            print("X. Xoay")
            print("P. Phóng to")
            print("H. Thu nhỏ")
            print("C. Coordinate Map")
            print("M. Quay lại menu chọn ảnh")
            print("0. Thoát chương trình")

            action_choice = input("Chọn thao tác (T/X/P/H/C/M/0): ").strip().upper()

            if action_choice == '0':
                print("Thoát chương trình. Tạm biệt!")
                return 
            elif action_choice == 'M':
                break 
            
            processed_image = None

            if action_choice == 'T':
                try:
                    shift_x = int(input("Nhập độ dịch chuyển theo chiều ngang (X): "))
                    shift_y = int(input("Nhập độ dịch chuyển theo chiều dọc (Y): "))
                    processed_image = apply_translation(current_image_data, shift_x, shift_y)
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số nguyên.")
            elif action_choice == 'X':
                try:
                    angle = float(input("Nhập góc xoay (độ): "))
                    reshape_choice = input("Thay đổi kích thước ảnh để chứa toàn bộ ảnh đã xoay? (Y/N): ").strip().upper()
                    reshape = True if reshape_choice == 'Y' else False
                    processed_image = apply_rotation(current_image_data, angle, reshape)
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số.")
            elif action_choice == 'P':
                try:
                    factor = float(input("Nhập hệ số phóng to (ví dụ: 2 cho gấp đôi): "))
                    if factor <= 1:
                        print("Hệ số phóng to phải lớn hơn 1.")
                        continue
                    processed_image = apply_zoom(current_image_data, factor)
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số.")
            elif action_choice == 'H':
                try:
                    factor = float(input("Nhập hệ số thu nhỏ (ví dụ: 0.5 cho giảm một nửa): "))
                    if factor >= 1 or factor <= 0:
                        print("Hệ số thu nhỏ phải nằm trong khoảng (0, 1).")
                        continue
                    processed_image = apply_zoom(current_image_data, factor)
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số.")
            elif action_choice == 'C':
                try:
                    d_val = int(input("Nhập biên độ dịch chuyển ngẫu nhiên (d, ví dụ: 5-20): "))
                    processed_image = apply_coordinate_map(current_image_data, d=d_val)
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số nguyên.")
            else:
                print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
                continue

            if processed_image is not None:
                plt.imshow(processed_image)
                plt.title(f"Kết quả {action_choice} trên {selected_image_path}")
                plt.axis('off')
                plt.show()

if __name__ == "__main__":
    main_menu()
```
